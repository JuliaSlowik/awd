```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
dane = pd.read_csv("vgsales.csv")
dane
```

```python
dane = dane.dropna()
dane = dane.drop_duplicates()
dane
```

```python
dane.drop(columns = 'Rank', inplace = True)
#wyÅ¼ej usuwam kolumnÄ™ "Rank", poniewaÅ¼ stanowi jedynie kopiÄ™ indeksowania wierszÃ³w, jest efektem #sortowania zbioru pod wzglÄ™dem sprzedaÅ¼y globalnej i raczej nie bÄ™dziemy jej uÅ¼ywaÄ‡
dane
```

```python
dane['Year'] = dane['Year'].astype(int)
dane['Platform'] = dane['Platform'].astype('category')
dane['Genre'] = dane['Genre'].astype('category')
dane['Publisher'] = dane['Publisher'].astype('category')
dane.info()
```

```python
lata_act=dane[dane.Genre=='Action'].groupby('Year').sum()
lata_plat=dane[dane.Genre=='Platform'].groupby('Year').sum()
lata_act
```

```python
act_dict=dict(lata_act.loc[:,'Global_Sales'])
plat_dict = dict(lata_plat.loc[:,'Global_Sales'])
plt.plot(act_dict.keys(),act_dict.values(),label='Action')
plt.plot(plat_dict.keys(),plat_dict.values(),label='Platform')
#kiedy Year jest objektem:
# plt.xticks([0,len(act_dict.keys())//2,len(act_dict.keys())-1])
#kiedy Year jest intem
#plt.xticks(rotation=80)
plt.title('SprzedaÅ¼ gier akcji i platformowych w latach 1980-2017')
plt.xlabel('Rok [-]')
plt.ylabel('Liczba sprzedanych egzemplarzy [-]')
#plt.xlim([1979,2016])
plt.legend()
plt.show()
```

```python
t,p = st.ttest_ind(lata_act['Global_Sales'],lata_plat['Global_Sales'])
print('t =',round(t,5),'\np =',round(p,5))
print(lata_act['Global_Sales'].mean(),lata_plat['Global_Sales'].mean(),sep='\n')
```

```python
pub=dane.Publisher.value_counts()
pub
```

```python
pub = dict(pub)
plt.bar(list(pub.keys())[:20],list(pub.values())[:20])
plt.title('Liczba gier poszczegÃ³lnych wydawcÃ³w')
plt.xlabel('Wydawca')
plt.ylabel('Liczba gier')
plt.xticks(rotation=90)
plt.show()
```

```python
gen = dane.Genre.value_counts()
gen_dict = dict(gen)
plt.bar(gen_dict.keys(),gen_dict.values())
plt.xticks(rotation=70)
plt.xlabel('Gatunek')
plt.ylabel('Liczba gier')
plt.title('Liczba gier dla poszczegÃ³lnych gatunkÃ³w')
```

```python
plat = dane.Platform.value_counts()
plat = dict(plat)
plat_istotne = {k:v for k,v in plat.items() if v>9}
plt.bar(plat_istotne.keys(),plat_istotne.values())
plt.title('Liczba gier wydanych na poszczegÃ³lne platformy')
plt.xlabel('Platforma')
plt.ylabel('Liczba gier')
plt.xticks(rotation=90)
```
